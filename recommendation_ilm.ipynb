```python
import pandas as pd
import random

# Генерация случайных данных
def generate_random_data():
    names = ["Алексей", "Мария", "Дмитрий", "Анна", "Иван", "Елена", "Николай", "Ольга", "Сергей", "Татьяна", "Петр", "Юлия"]
    genders = ["Мужской", "Женский"]
    classes = list(range(5, 12))
    subjects = ["Русский", "Английский", "Физика", "Химия", "Биология", "Право", "Физкультура", "Геометрия", "Алгебра", "Информационные технологии", "Основы экономики", "География", "История таджикского народа", "Всеобщая история", "Всемирная литература"]
    languages = ["Таджикский", "Русский", "Английский"]
    study_methods = ["Теоретический"]
    interests = ["Наука", "Спорт", "Искусство", "Музыка", "Технологии", "Литература", "История", "Экономика"]
    goals = ["Поступить в университет", "Получить стипендию", "Углубить знания"]
    health_statuses = ["Здоров", "Хронические болезни", "Инвалидность"]

    data = {
        "ID": [i+1 for i in range(1000)],
        "Имя": [random.choice(names) for _ in range(1000)],
        "Возраст": [random.randint(11, 18) for _ in range(1000)],
        "Пол": [random.choice(genders) for _ in range(1000)],
        "Класс": [random.choice(classes) for _ in range(1000)],
        "Имя предмета": [random.choice(subjects) for _ in range(1000)],
        "Изучаемый язык": [random.choice(languages) for _ in range(1000)],
        "Метод обучения": [random.choice(study_methods) for _ in range(1000)],
        "Интерес": [random.choice(interests) for _ in range(1000)],
        "Цель": [random.choice(goals) for _ in range(1000)],
        "Здоровье": [random.choice(health_statuses) for _ in range(1000)]
    }

    return data

# Создание DataFrame
data = generate_random_data()
df = pd.DataFrame(data)

# Сохранение в CSV файл
df.to_csv("students_database.csv", index=False)
```

```python
 import pandas as pd
import random
from datetime import datetime, timedelta

# Существующая база данных студентов
students_data = {
    "ID": [1, 2, 3, 4, 5],
    "Имя": ["Николай", "Татьяна", "Дмитрий", "Анна", "Мария"],
    "Возраст": [12, 17, 12, 14, 16],
    "Пол": ["Мужской", "Женский", "Мужской", "Женский", "Женский"],
    "Класс": [5, 9, 8, 9, 7],
    "Имя предмета": ["Физика", "Всемирная литература", "Физика", "Всемирная литература", "Право"],
    "Изучаемый язык": ["Таджикский", "Таджикский", "Таджикский", "Русский", "Английский"],
    "Метод обучения": ["Теоретический", "Теоретический", "Теоретический", "Теоретический", "Теоретический"],
    "Интерес": ["История", "Технологии", "История", "Литература", "Спорт"],
    "Цель": ["Поступить в университет", "Поступить в университет", "Поступить в университет", "Углубить знания", "Углубить знания"],
    "Здоровье": ["Инвалидность", "Здоров", "Хронические болезни", "Инвалидность", "Хронические болезни"]
}

students_df = pd.DataFrame(students_data)

# Генерация таблицы рекомендаций
def generate_recommendations(num_recommendations, students_df):
    types_of_recommendations = ["Учебный материал", "Дополнительные упражнения", "Ресурсы", "Методы обучения"]
    status_options = ["Просмотрено", "Выполнено", "Не актуально"]
    algorithms = ["Алгоритм 1", "Алгоритм 2", "Алгоритм 3"]

    recommendations = {
        "Идентификатор студента": [],
        "Тип рекомендации": [],
        "Описание рекомендации": [],
        "Основа рекомендации": [],
        "Дата рекомендации": [],
        "Статус": []
    }

    for _ in range(num_recommendations):
        student_id = random.choice(students_df["ID"].tolist())
        recommendations["Идентификатор студента"].append(student_id)
        recommendations["Тип рекомендации"].append(random.choice(types_of_recommendations))
        recommendations["Описание рекомендации"].append("Описание материала для студента ID {}".format(student_id))
        recommendations["Основа рекомендации"].append(random.choice(algorithms))
        recommendations["Дата рекомендации"].append(datetime.now() - timedelta(days=random.randint(0, 365)))
        recommendations["Статус"].append(random.choice(status_options))

    return pd.DataFrame(recommendations)

# Создание таблицы рекомендаций
recommendations_df = generate_recommendations(20, students_df)

# Сохранение в CSV файл (если нужно)
students_df.to_csv("students_database.csv", index=False)
recommendations_df.to_csv("recommendations_database.csv", index=False)
```
